```python
df = pandas.read_excel('s3://lowe-241-2023-python/covid_data.xls')
df[:5]
```

## Lithuania Data

```python
Lithuania_data= df.loc[df['countriesAndTerritories']=='Lithuania']

Lithuania_data[:5]
```

## Question 1: What was the Month with the Highest Spike in Covid Cases?

```python
Lithuania_cases_by_month = Lithuania_data.groupby('month')['cases'].sum()

Lithuania_cases_by_month.plot() 
```

## Interpretation and Explantation 

This line graph depicts the number of Covid cases per month in Lithuania. In November of 2020, Lithuania saw its highest spike in cases with over 45,000 confirmed cases. 

## Question 2: What was the Month with the Highest Number of Deaths?

```python
Lithuania_deaths_by_month = Lithuania_data.groupby('month')['deaths'].sum()

Lithuania_deaths_by_month.plot()
```

## Interpretation and Explantation 

This line graph depicts the number of Covid deaths per month in Lithuania. In November of 2020, Lithuania saw the highest amount of deaths, about 350 people died. 

## Denmark Data

## Question 3: The Number of Cases in Lithuania Compared to Denmark
